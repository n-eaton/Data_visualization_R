```r
gapminder %>%
    select(country, lifeExp) %>%
    filter(country == "Bulgaria" |
          country == "Brazil" |
          country == "China" |
          country == "Egypt"|
          country == "Uganda") %>%
 group_by(country) %>%
summarise(Average_life = mean(lifeExp))
```

#### t-test

```r
df1 <-gapminder %>%
    select(country, lifeExp) %>%
    filter(country == "Bulgaria" |
          country == "Brazil" ) 
t.test(data=df1, lifeExp ~ country)
```

```r
ggplot(subset(gapminder, continent != "Oceania"),
       aes(x = year, y = lifeExp, group = country, color = country)) +
  geom_line(lwd = 1, show.legend = FALSE) + facet_wrap(~ continent) +
  scale_color_manual(values = country_colors) +
  theme_bw() + theme(strip.text = element_text(size = rel(1.1)))
```

```r
gapminder %>%
    group_by(continent, year) %>%
    summarise(lifeExp=median(lifeExp)) -> gapyear
```

```r
ggplot(gapyear, aes(x=year, y=lifeExp, color=continent)) +
    geom_point(size=1.5) +
    geom_smooth(aes(fill=continent), method="lm")
```

```r
gap_with_colors <-
  data.frame(gapminder,
             cc = I(country_colors[match(gapminder$country,
                                         names(country_colors))]))
```

```r
keepers <- with(gap_with_colors,
                continent %in% c("Africa", "Oceania") & year == 2007)
plot(lifeExp ~ gdpPercap, gap_with_colors,
     subset = keepers, log = "x", pch = 21,
     cex = sqrt(gap_with_colors$pop[keepers]/pi)/1500,
     bg = gap_with_colors$cc[keepers])
```
